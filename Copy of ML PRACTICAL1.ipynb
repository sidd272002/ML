```python
df['Date']=pd.to_datetime(df['date_issued'])
```

```python
df['Month']= df['Date'].dt.month
```

```python
df['date:Year']=df['Date'].dt.year
```

```python
df['Month']
```

```python
df['date:month']=df['Date'].dt.month
```

```python
df['date:day']=df['Date'].dt.day
```

```python
df['date:week']=df['Date'].dt.day_of_year
```

```python
df['day:Year']=df['Date'].dt.day_of_year
```

```python
df[['date:month','date:day','date:week','day:Year']]
```

```python
def week_part(day):
  if day in [1,2,3,4,5,6,7]:
    return ('week1')   
  elif day in[8,9,10,11,12,13,14]:
    return ("week2")
  elif day in[15,26,17,18,19,20,21]:
    return ("week3")
  elif day in[22,23,24,25,26,27,28]:
    return ("week4")
  elif day in[29,30,31]:
    return ("week5")
```

```python
df['date:week_no']=df['date:day'].apply(week_part)
```

```python
df.head(10)
```

```python
df[['date:day','date:week_no']]
```

```python
df['is_weekend']=np.where(df['date:week'].isin([0,6]),1,0)
```

```python
df[['Date','date:week','is_weekend']]
```
