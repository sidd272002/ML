```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_csv("iphone_purchase_records(1).csv")
```

```python
dataset.head()
```

```python
dataset['log_Age']=np.log(dataset['Age'])
dataset['log_sal']=np.log(dataset['Salary'])
dataset.head()
```

```python
x=np.array(dataset[['Gender','log_Age','log_sal']])
y=np.array(dataset['Purchase Iphone'])

from sklearn.preprocessing import LabelEncoder

label_Encoder_gender = LabelEncoder()
x[:,0]=label_Encoder_gender.fit_transform(x[:,0])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
ss_x = StandardScaler()
x_train = ss_x.fit_transform(x_train)
x_test = ss_x.transform(x_test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel="linear",random_state=0)
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
```

```python
from sklearn import metrics 
cm = metrics.confusion_matrix(y_test,y_pred)
print(cm)
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
print("Accuracy Score:",accuracy)
precision = metrics.precision_score(y_test,y_pred)
print("Precision Score:",precision)
```

```python
recall=metrics.recall_score(y_test,y_pred)
print("Recall Score:",recall)
```
